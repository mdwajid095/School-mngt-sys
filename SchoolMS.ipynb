```python
import sqlite3
from tkinter import *

try:
    db=sqlite3.connect("CA2017.db")
    cr=db.cursor()
    cr.execute("create table student(Roll integer primary key ,Name text, Gender text,  Class text )")
    db.commit()
except:
    print('table is available')
    

w=Tk()
w.geometry('600x500')
w.configure(background="skyblue")
w.title("CA DATABASE")

p=StringVar()
q=StringVar()
r=StringVar()
s=StringVar()

L1=Label(w,text="Career Amplifier", font=("none 38 bold") , bg='skyblue', fg="blue")
L1.pack()

L1=Label(w,text="Roll : ", font=("none 14 bold") , bg='skyblue', fg="black")
L1.place(x=100,y=110)
E1=Entry(w, font=("none 14 bold"), fg="black", bg='cyan', textvar=p)
E1.place(x=185,y=110)


L2=Label(w,text="Name:", font=("none 14 bold") , bg='skyblue', fg="black")
L2.place(x=100,y=150)
E2=Entry(w, font=("none 14 bold"), fg="black",bg='cyan', textvar=q)
E2.place(x=185,y=150)
          
    
L3=Label(w,text="Class : ", font=("none 14 bold") , bg='skyblue', fg="black")
L3.place(x=100,y=190)
E3=Entry(w, font=("none 14 bold"), fg="black", bg='cyan', textvar=r)
E3.place(x=185,y=190)


L4=Label(w,text="Gender: ", font=("none 14 bold") , bg='skyblue', fg="black")
L4.place(x=100,y=230)
s=StringVar()
Radiobutton(w, text="Male", padx=10, bg='skyblue', variable=s,value='Male').place(x=190,y=230)
Radiobutton(w, text="Female", padx=20,bg='skyblue', variable=s,value='Female').place(x=280,y=230)


def insert():
    a=p.get()
    b=q.get()
    c=r.get()
    d=s.get()
    cr.execute("insert into student(Roll,Name,Class,Gender)values(?,?,?,?)", (a,b,c,d))
    db.commit()
    
    
Bins=Button(text="Submit", fg="blue" , padx=10,pady=3, font=("none 14 bold") ,command= insert)
Bins.place(x=40,y=280)


def delete():
    a=p.get()
    cr.execute("delete from student where Roll=?",(a,))
    db.commit()
    
Bdel=Button(text='Delete',fg='blue',padx=12,pady=7,font=('none 13 bold'),command=delete)
Bdel.place(x=150,y=280)

def update():
    a=p.get()
    b=q.get()
    c=r.get()
    d=s.get()
    cr.execute("update student set Name=?,Class=?,Gender=? where Roll=?",(b,c,d,a))
    db.commit()
    
Bupdt=Button(text='Update',fg='blue',padx=12,pady=7,font=('none 13 bold'),command=update)
Bupdt.place(x=240,y=280)


def fetch():
    a=p.get()
    cr.execute("select * from student where Roll=?",(a,))
    data=cr.fetchall()
    listNodes.delete(0,END)
    for i in data:
        listNodes.insert(END,i)
        
Bfetch=Button(text='Show',fg='blue',padx=12,pady=7,font=('none 13 bold'),command=fetch)
Bfetch.place(x=340,y=280)

def fetchall():
    cr.execute("select * from student order by Roll")
    data=cr.fetchall()
    listNodes.delete(0,END)
    for i in data:
        listNodes.insert(END,i)
        
Bfetch=Button(text='Show All',fg='blue',padx=12,pady=7,font=('none 11 bold'),command=fetchall)
Bfetch.place(x=430,y=280)


frame = Frame(w)
frame.place(x=100,y=350)
listNodes = Listbox(frame, width=40, height=6, font=("Helvetica", 12))
listNodes.pack(side="left", fill="y")

scrollbar = Scrollbar(frame, orient="vertical")
scrollbar.config(command=listNodes.yview)
scrollbar.pack(side="right", fill="y")

listNodes.config(yscrollcommand=scrollbar.set)

def get_selected_row(event):
    global selected_tuple
    if len(listNodes.curselection())>0:
        index=listNodes.curselection()[0]
        selected_tuple=listNodes.get(index)
        E1.delete(0,END)
        E1.insert(END,selected_tuple[1])
        E2.delete(0,END)
        E2.insert(END,selected_tuple[2])
        E1.delete(0,END)
        E1.insert(END,selected_tuple[3])
        E2.delete(0,END)
        E2.insert(END,selected_tuple[4])
        
listNodes.bind('<<ListboxSelect>>',get_selected_row)

w.mainloop()
```
